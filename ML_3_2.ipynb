```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("diabetes-3.csv")
```

```python
df.shape
```

```python
print("Data types:")
print(df.dtypes)

```

```python
print("\nMissing values per column:")
print(df.isnull().sum())

```

```python
print("\nDescriptive statistics for numerical columns:")
print(df.describe())

```

```python
for col in ['County', 'City', 'Postal Code', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Electric Range', 'Base MSRP', 'Legislative District', 'DOL Vehicle ID', 'Vehicle Location', 'Electric Utility', '2020 Census Tract']:
    if df[col].dtype == 'object':

        df[col].fillna(df[col].mode()[0], inplace=True)
    else:

        df[col].fillna(df[col].mean(), inplace=True)

print("\nMissing values after handling:")
print(df.isnull().sum())
```

```python
y = df["Electric Vehicle Type"]

X = df.drop(columns=["Electric Vehicle Type", "Postal Code"])
```

```python
from sklearn.metrics import roc_auc_score, make_scorer
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
import numpy as np

dt_model = DecisionTreeClassifier(random_state=42)

n_splits = 5

skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision_macro',
    'recall': 'recall_macro',
    'f1': 'f1_macro',
}

cv_results = {}
for metric_name, scorer in scoring.items():
    cv_scores = cross_val_score(dt_model, X, y, cv=skf, scoring=scorer)
    cv_results[metric_name] = cv_scores

for metric_name, scores in cv_results.items():
    print(f"{metric_name.capitalize()} CV scores: {scores}")
    print(f"Mean {metric_name.capitalize()} CV score: {np.mean(scores):.4f}\n")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier

dt_model_viz = DecisionTreeClassifier(random_state=42)
dt_model_viz.fit(X, y)


plt.figure(figsize=(20,10))
plot_tree(dt_model_viz, feature_names=X.columns, filled=True, max_depth=4)
plt.title("Decision Tree Structure (max_depth=3)")
plt.show()
```
