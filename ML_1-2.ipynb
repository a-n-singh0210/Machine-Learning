```python
df['Value'] = df['Value'].fillna(df['Value'].median())
df['LowCI'] = df['LowCI'].fillna(df['LowCI'].median())
df['HighCI'] = df['HighCI'].fillna(df['HighCI'].median())
```

```python
df.isnull().sum()
```

```python


ci_values = []

for i in range(len(df)):
    low = round(df.loc[i, 'LowCI'], 1)
    high = round(df.loc[i, 'HighCI'], 1)
    ci_string = f"{low} - {high}"
    ci_values.append(ci_string)

df['Confidence Interval'] = ci_values
```

```python
#df['Confidence Interval'] = df['Confidence Interval'].fillna(df['Confidence Interval'].mode()[0]) #1st mode
df['Quartile Range'] = df['Quartile Range'].fillna(df['Quartile Range'].mode()[0])
```

```python
df['Time Period Start Date'] = pd.to_datetime(df['Time Period Start Date'])
df_time_trends = df.groupby('Time Period Start Date')[['Value', 'LowCI', 'HighCI']].mean()

plt.figure(figsize=(12, 6))
plt.plot(df_time_trends.index, df_time_trends['Value'])
plt.title('Mean Value Over Time')
plt.xlabel('Time Period Start Date')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_state = df[df['Group'] == 'By State']
df_state_avg = df_state.groupby('State')[['Value', 'LowCI', 'HighCI']].mean().reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=df_state_avg, x='State', y='Value')
plt.title('Mean Value of Mental Health Indicators by State')
plt.xlabel('State')
plt.ylabel('Mean Value')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
# Plot mean Value by Indicator
plt.figure(figsize=(12, 6))
df_indicator_avg = df.groupby('Indicator')['Value'].mean().sort_values()
df_indicator_avg.plot(kind='barh')
plt.title('Mean Value of Mental Health Indicators')
plt.xlabel('Mean Value')
plt.ylabel('Indicator')
plt.tight_layout()
plt.show()

# Plot mean Value by Subgroup within each Group
for group in df['Group'].unique():
    plt.figure(figsize=(12, 6))
    df_subgroup_avg = df[df['Group'] == group].groupby('Subgroup')['Value'].mean().sort_values()
    df_subgroup_avg.plot(kind='barh')
    plt.title(f'Mean Value of Mental Health Indicators by Subgroup ({group})')
    plt.xlabel('Mean Value')
    plt.ylabel('Subgroup')
    plt.tight_layout()
    plt.show()
```
