```python
# Confidence Interval	ke liye MODE
#Quartile Range
```

```python
df['Value'] = df['Value'].fillna(df['Value'].mean())
df['LowCI'] = df['LowCI'].fillna(df['LowCI'].mean())
df['HighCI'] = df['HighCI'].fillna(df['HighCI'].mean())
```

```python
df.isnull().sum()
```

```python
df['Confidence Interval'] = df['Confidence Interval'].fillna(df['Confidence Interval'].mode()[0]) #1st mode
df['Quartile Range'] = df['Quartile Range'].fillna(df['Quartile Range'].mode()[0])
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

numerical_columns = ['Value', 'LowCI', 'HighCI']

for col in numerical_columns:
    df.boxplot(column=col)
    plt.title(f"Box plot of {col}")
    plt.show()
```

```python
numerical_df = df[['Value', 'LowCI', 'HighCI']]
correlation_matrix = numerical_df.corr()
display(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
